```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
data=np.genfromtxt('/Users/dmelgar/Classes/GEOL410/data/sst_mac.txt')
lon=data[:,0]
lat=data[:,1]
sst=data[:,2]
```

```python
plt.figure()
plt.scatter(lon,lat,s=0.2)
plt.show()
```

```python
plt.figure()
plt.plot(lon,sst)
plt.show()
```

```python
X,Y=np.meshgrid(np.unique(lon),np.unique(lat))
print(X.shape)
print(Y.shape)

print(X)
print(Y)

Z=np.zeros(X.shape)
print(Z)
```

```python
plt.figure()
plt.scatter(lon,lat,c=sst)
plt.show()
```

```python
k=0
for klat in range(len(np.unique(lat))):
    for klon in range(len(np.unique(lon))):
        Z[klat,klon]=sst[k]
        k+=1
```

```python
plt.figure()
plt.contour(X,Y,Z,40,linewidths=0.8,cmap='magma')
cb=plt.colorbar()
cb.set_label('SST (deg. C)')
plt.show()
```

```python
plt.figure()
contour=plt.contour(X,Y,Z,10,linewidths=0.8,cmap='rainbow')
plt.clabel(contour, inline=1, fontsize=10)
cb=plt.colorbar()
cb.set_label('SST (deg. C)')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.pcolor(X,Y,Z,cmap='jet')
cb=plt.colorbar()
cb.set_label('SST (deg. C)')
contour=plt.contour(X,Y,Z,10,linewidths=0.8,cmap='jet')
plt.clabel(contour, inline=1, fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.pcolor(X,Y,Z,cmap='brg')
cb=plt.colorbar()
cb.set_label('SST (deg. C)')
contour=plt.contour(X,Y,Z,10,linewidths=0.8,colors='k')
plt.clabel(contour, inline=1, fontsize=10)
plt.show()
```
