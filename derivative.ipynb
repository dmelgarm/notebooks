## Derivative of a complciated function

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook

```

```python
x=np.arange(0,50,0.1)
print(x)
```

```python
y=(3*x**2*(np.sin(x))-x**3)/((x+1)**3)
```

```python
plt.figure()
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
derivative=np.zeros(len(x))
deltax=x[1]-x[0]
for k in range(len(x)-1):
    derivative[k+1]=(y[k+1]-y[k])/deltax
```

```python
plt.figure()
plt.plot(x,y)
plt.plot(x,derivative)
plt.legend(['y','dy/dx'])
plt.grid()
plt.show()
```
