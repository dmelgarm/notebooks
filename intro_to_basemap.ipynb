```python
data=np.genfromtxt('/Users/dmelgar/Classes/GEOL410/data/sst_mac.txt')
lon=data[:,0]
lat=data[:,1]
sst=data[:,2]

X,Y=np.meshgrid(np.unique(lon),np.unique(lat))
Z=np.zeros(X.shape)
k=0
for klat in range(len(np.unique(lat))):
    for klon in range(len(np.unique(lon))):
        Z[klat,klon]=sst[k]
        
        k+=1
```

```python


plt.figure(figsize=(12,6))

m = Basemap(projection='robin',lon_0=0,resolution='l')
lons, lats = m(X, Y)
m.drawcoastlines()
m.fillcontinents(color='brown',lake_color='aqua')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.),labels=[False,True,False,False])
m.drawmeridians(np.arange(0.,360.,60.),labels=[False,False,False,True])
m.contour(lons,lats,Z,cmap='jet')
# m.drawmapboundary(fill_color='cyan')

plt.show()
```

```python
plt.figure(figsize=(12,6))

m = Basemap(projection='robin',lon_0=0,resolution='l')
lons, lats = m(X, Y)
m.drawcoastlines()
m.fillcontinents(color='brown',lake_color='aqua')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.),labels=[False,True,False,False])
m.drawmeridians(np.arange(0.,360.,60.),labels=[False,False,False,True])
m.contour(lons,lats,Z,30,cmap='jet')
# m.drawmapboundary(fill_color='cyan')

plt.show()
```

```python
Zmask=maskoceans(lonsin=X,latsin=Y,datain=Z)
Zmask_oceans = np.ma.masked_array(Z, mask=~Zmask.mask)

plt.figure(figsize=(12,6))

m = Basemap(projection='robin',lon_0=0,resolution='l')
lons, lats = m(X, Y)
m.drawcoastlines()
m.fillcontinents(color='brown',lake_color='aqua')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.),labels=[False,True,False,False])
m.drawmeridians(np.arange(0.,360.,60.),labels=[False,False,False,True])
m.pcolor(lons,lats,Zmask_oceans,cmap='jet')

plt.show()
```
