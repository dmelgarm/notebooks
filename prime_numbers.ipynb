## Is a number prime?

This notebook will attempt to answer the afore mentioned question with a simple algorithm that takes tries to see if the candidate numebr has any divisors

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from scipy import stats
%matplotlib notebook
import timeit
```

First we're going to define the potential divisors of a test number

```python
start = timeit.default_timer()

test_number=2000
potential_divisors=np.arange(2,test_number-1)
# print(potential_divisors)
```

Now we're going to loop over all ther possible divisors and see if they are or aren't a divisor

```python
is_number_prime=False

for divisor in potential_divisors:
    print("Currently testing potential divisor: "+str(divisor))
    remainder=test_number%divisor
    
    #if remainder is zero the the number is a divisor so make the is_number_prime be true
    if remainder == 0:
        print('Aha! Found a divisor '+str(test_number)+' is divisble by '+str(divisor))
        is_number_prime=True
        
```

Now if no divisor was found then inform the user

```python
if is_number_prime==False:
    print("I couldn't fin any divisors so the number must be prime, congrats")
```

```python
stop = timeit.default_timer()
duration=stop-start
print("It took me "+str(duration)+"seconds to run")
```

```python
def is_prime(test_number):
    
    potential_divisors=np.arange(2,test_number-1)
    is_number_prime=True

    for divisor in potential_divisors:
        remainder=test_number%divisor

        #if remainder is zero the the number is a divisor so make the is_number_prime be true
        if remainder == 0:
            is_number_prime=False
            
    return is_number_prime
```

```python
max_prime=10000
test_numbers=np.arange(3,max_prime)
results=np.zeros(len(test_numbers))
```

```python
for k in range(len(test_numbers)):
    was_number_prime=is_prime(test_numbers[k])
    if was_number_prime==True:
        results[k]=1
```

```python
# plt.figure()
# plt.stem(test_numbers,results)
```

```python
bins=np.arange(0,max_prime,100)
count,edges,numbers=stats.binned_statistic(test_numbers,results,bins=bins,statistic='sum')
print(bins)
print(count)
```

```python
plt.figure()
plt.bar(edges[0:-1],count,width=100)
plt.xlabel('Bin center')
plt.ylabel('# of primes in bin')
```
