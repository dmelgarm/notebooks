```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
data=np.genfromtxt('/Users/dmelgar/Classes/GEOL410/data/mauna_loa_weekly_raw.csv')
```

```python
dt=7
days=np.arange(0,len(data)*dt,dt)
co2=data[:,1]
```

```python
plt.figure()
plt.scatter(days,co2)
plt.show()
```

```python
plt.figure()
plt.scatter(days,co2,s=1,marker='s')
plt.show()
```

```python
plt.figure()
plt.plot(days,co2)
plt.show()
```

```python
plt.figure()
plt.plot(days,co2)
plt.xlabel('Days since 1958-03-29')
plt.ylabel('CO2 (ppm)')
plt.show()
```

```python
years=days/365
```

```python
plt.figure()
plt.plot(years,co2)
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 (ppm)')
plt.show()
```

```python
plt.figure()
plt.plot(years,co2)
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 (ppm)')
plt.ylim([0,450])
plt.grid()
plt.show()
```

```python
rate=np.diff(co2)

plt.figure()

plt.subplot(2,1,1)
plt.plot(years,co2)
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 (ppm)')

plt.subplot(2,1,2)
plt.plot(years[1:],rate)
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 change (ppm/week)')

plt.show()
```

```python
rate=np.diff(co2)

plt.figure()

plt.subplot(2,1,1)
plt.plot(years,co2)
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 (ppm)')

plt.subplot(2,1,2)
plt.plot(years[1:],rate)
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 change (ppm/week)')

plt.subplots_adjust(hspace=0.5)  #or 0

#remove xlabel on upper panel

plt.show()
```

```python
rate=np.diff(co2)

plt.figure()

ax1=plt.subplot(2,1,1)
ax1.plot(years,co2)
ax1.set_xlabel('Years since 1958-03-29')
ax1.set_ylabel('CO2 (ppm)')

ax2=plt.subplot(2,1,2,sharex=ax1)
ax2.plot(years[1:],rate)
ax2.set_xlabel('Years since 1958-03-29')
ax2.set_ylabel('CO2 change (ppm/week)')

plt.show()
```

```python
#draw a reference line

#yreference=0.1*years+320
yreference=1*years+314

rate=np.diff(co2)

plt.figure()

plt.plot(years,co2)
plt.plot(years,yreference)
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 (ppm)')
plt.grid()

plt.legend(['Measurements','Reference trend'])



plt.show()
```

```python
yreference=1*years+314

rate=np.diff(co2)

plt.figure()

plt.fill_between(years,co2,yreference,color='red',label='Deviation from trend',alpha=0.2)
plt.plot(years,co2,label='Measurements')
plt.plot(years,yreference,c='#606060',label='Reference trend')
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 (ppm)')
plt.grid()
plt.legend()



plt.show()
```

```python
yreference=1*years+314

rate=np.diff(co2)

plt.figure()

plt.fill_between(years,co2,yreference,color='red',label='Deviation from trend',alpha=0.4)
plt.plot(years,co2,label='Measurements')
plt.plot(years,yreference,c='#606060',label='Reference trend')
plt.plot([0,60],[400,400],lw=2,c='r')
plt.xlabel('Years since 1958-03-29')
plt.ylabel('CO2 (ppm)')
plt.grid()
plt.legend()

plt.annotate(s='Never exceeded in last 4,000,000 yrs (yay!)',xy=(1,402))



plt.show()
```
